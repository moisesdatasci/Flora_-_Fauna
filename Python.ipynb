```python
# ==========================================
# PROYECTO: FLORA Y FAUNA DE CHILE
# Autor: MOISÃ‰S JOEL ORTEGA CATRIÃ‘IR
# Fecha: 2025
# ==========================================

# 1. CONCEPTOS FUNDAMENTALES DE PYTHON
# Variables bÃ¡sicas con diferentes tipos de datos
nombre_programa = "Flora y Fauna de Chile"
version = 1.0
total_especies = 18
programa_activo = True

# 2. ESTRUCTURAS DE DATOS - DICCIONARIO PRINCIPAL
# Base de datos simple con flora y fauna por regiÃ³n
biodiversidad = {
    "Norte": {
        "flora": ["QueÃ±oa", "ChaÃ±ar", "Yareta"],
        "fauna": ["VicuÃ±a", "Flamenco", "Chinchilla"]
    },
    "Centro": {
        "flora": ["Palma Chilena", "Espino", "Quillay"],
        "fauna": ["PudÃº", "CÃ³ndor", "Zorro Culpeo"]
    },
    "Sur": {
        "flora": ["Araucaria", "CoigÃ¼e", "ArrayÃ¡n"],
        "fauna": ["Huemul", "Puma", "Monito del Monte"]
    }
}

# Lista de regiones disponibles
regiones = ["Norte", "Centro", "Sur"]

# ==========================================
# 3. FUNCIONES PARA REUTILIZACIÃ“N DE CÃ“DIGO
# ==========================================

def mostrar_menu():
    """FunciÃ³n que muestra el menÃº principal"""
    print("\n" + "="*40)
    print(f"ðŸŒ¿ {nombre_programa} v{version}")
    print("="*40)
    print("1. Ver flora por regiÃ³n")
    print("2. Ver fauna por regiÃ³n")
    print("3. Buscar especie")
    print("4. Agregar nueva especie")
    print("5. Contar especies")
    print("6. Salir")
    print("="*40)

def mostrar_flora(region):
    """
    FunciÃ³n para mostrar flora de una regiÃ³n
    ParÃ¡metro: region (string)
    """
    print(f"\nðŸŒ± FLORA DE LA REGIÃ“N {region.upper()}:")
    print("-" * 30)

    # Bucle FOR para mostrar cada planta
    for i, planta in enumerate(biodiversidad[region]["flora"], 1):
        print(f"{i}. {planta}")

def mostrar_fauna(region):
    """
    FunciÃ³n para mostrar fauna de una regiÃ³n
    ParÃ¡metro: region (string)
    """
    print(f"\nðŸ¦Ž FAUNA DE LA REGIÃ“N {region.upper()}:")
    print("-" * 30)

    # Bucle FOR para mostrar cada animal
    for i, animal in enumerate(biodiversidad[region]["fauna"], 1):
        print(f"{i}. {animal}")

def contar_especies():
    """
    FunciÃ³n para contar especies por regiÃ³n
    Retorna: diccionario con conteos
    """
    conteos = {}

    # Bucle FOR para recorrer regiones
    for region in regiones:
        flora_count = len(biodiversidad[region]["flora"])
        fauna_count = len(biodiversidad[region]["fauna"])
        total = flora_count + fauna_count

        conteos[region] = {
            "flora": flora_count,
            "fauna": fauna_count,
            "total": total
        }

    return conteos

def buscar_especie(nombre_buscar):
    """
    FunciÃ³n para buscar una especie en todas las regiones
    ParÃ¡metro: nombre_buscar (string)
    Retorna: True si encuentra, False si no
    """
    encontrado = False

    # Bucles FOR anidados para buscar
    for region in regiones:
        for tipo in ["flora", "fauna"]:
            for especie in biodiversidad[region][tipo]:
                # Sentencia IF para comparar
                if nombre_buscar.lower() in especie.lower():
                    print(f"âœ… Encontrado: {especie} - {tipo} de regiÃ³n {region}")
                    encontrado = True

    return encontrado

def agregar_especie(region, tipo, nombre):
    """
    FunciÃ³n para agregar nueva especie
    ParÃ¡metros: region, tipo, nombre (strings)
    """
    # Sentencia IF para validar regiÃ³n
    if region in regiones:
        # Sentencia IF para validar tipo
        if tipo in ["flora", "fauna"]:
            biodiversidad[region][tipo].append(nombre)
            print(f"âœ… {nombre} agregado a {tipo} de {region}")
            return True
        else:
            print("âŒ Tipo debe ser 'flora' o 'fauna'")
            return False
    else:
        print("âŒ RegiÃ³n no vÃ¡lida")
        return False

def obtener_numero(mensaje, minimo, maximo):
    """
    FunciÃ³n para obtener nÃºmero vÃ¡lido del usuario
    Manejo bÃ¡sico de errores
    """
    # Bucle WHILE para validar entrada
    while True:
        try:
            numero = int(input(mensaje))
            # Sentencia IF para validar rango
            if minimo <= numero <= maximo:
                return numero
            else:
                print(f"âŒ Ingrese un nÃºmero entre {minimo} y {maximo}")
        except:
            print("âŒ Por favor ingrese solo nÃºmeros")

# ==========================================
# 4. PROGRAMA PRINCIPAL
# ==========================================

def main():
    """FunciÃ³n principal del programa"""
    print("ðŸŒ¿ Â¡Bienvenido al sistema de Flora y Fauna de Chile!")

    # Bucle WHILE principal
    while programa_activo:
        mostrar_menu()

        # Obtener opciÃ³n del usuario
        opcion = obtener_numero("Seleccione una opciÃ³n: ", 1, 6)

        # 5. SENTENCIAS CONDICIONALES
        if opcion == 1:
            # Ver flora por regiÃ³n
            print("\nREGIONES DISPONIBLES:")
            for i, region in enumerate(regiones, 1):
                print(f"{i}. {region}")

            region_num = obtener_numero("Seleccione regiÃ³n: ", 1, 3)
            region_elegida = regiones[region_num - 1]
            mostrar_flora(region_elegida)

        elif opcion == 2:
            # Ver fauna por regiÃ³n
            print("\nREGIONES DISPONIBLES:")
            for i, region in enumerate(regiones, 1):
                print(f"{i}. {region}")

            region_num = obtener_numero("Seleccione regiÃ³n: ", 1, 3)
            region_elegida = regiones[region_num - 1]
            mostrar_fauna(region_elegida)

        elif opcion == 3:
            # Buscar especie
            nombre = input("\nIngrese nombre de especie a buscar: ")

            # Sentencia IF para verificar resultado
            if not buscar_especie(nombre):
                print("âŒ No se encontrÃ³ la especie")

        elif opcion == 4:
            # Agregar nueva especie
            print("\nREGIONES DISPONIBLES:")
            for i, region in enumerate(regiones, 1):
                print(f"{i}. {region}")

            region_num = obtener_numero("Seleccione regiÃ³n: ", 1, 3)
            region_elegida = regiones[region_num - 1]

            print("\nTIPOS DISPONIBLES:")
            print("1. Flora")
            print("2. Fauna")

            tipo_num = obtener_numero("Seleccione tipo: ", 1, 2)
            tipo_elegido = "flora" if tipo_num == 1 else "fauna"

            nombre_especie = input("Ingrese nombre de la especie: ")

            # Llamar funciÃ³n para agregar
            agregar_especie(region_elegida, tipo_elegido, nombre_especie)

        elif opcion == 5:
            # Contar especies
            print("\nðŸ“Š ESTADÃSTICAS DE ESPECIES:")
            print("=" * 40)

            # Llamar funciÃ³n y mostrar resultados
            estadisticas = contar_especies()

            # Bucle FOR para mostrar estadÃ­sticas
            for region, datos in estadisticas.items():
                print(f"\nðŸŒ {region}:")
                print(f"  ðŸŒ± Flora: {datos['flora']}")
                print(f"  ðŸ¦Ž Fauna: {datos['fauna']}")
                print(f"  ðŸ“Š Total: {datos['total']}")

            # Calcular total general
            total_general = sum(datos['total'] for datos in estadisticas.values())
            print(f"\nðŸŒ TOTAL GENERAL: {total_general} especies")

        elif opcion == 6:
            # Salir del programa
            print("\nðŸŒ¿ Â¡Gracias por usar el sistema!")
            print("ðŸ¦Ž Â¡Cuida la flora y fauna de Chile!")
            break

        # Pausa para leer resultados
        input("\nPresione Enter para continuar...")


if __name__ == "__main__":
    main()
```
